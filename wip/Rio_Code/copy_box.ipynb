```python
def GetSol_dict (my_dict, keys_all, Sol, 
    cycle_no,step_CD, step_CC , step_RE, step_CV ):

    [keys_loc,keys_tim,keys_cyc]  = keys_all;
    # get time_based variables:
    if len(keys_tim): 
        for key in keys_tim:
            step_no = eval("step_{}".format(key[0:2]))
            if key[3:] == "Time [h]":
                my_dict[key].append  (
                    Sol.cycles[cycle_no].steps[step_no][key[3:]].entries
                    -
                    Sol.cycles[cycle_no].steps[step_no][key[3:]].entries[0])
            else:
                #print("Solve up to Step",step_no)
                my_dict[key].append  (
                    Sol.cycles[cycle_no].steps[step_no][key[3:]].entries)
    # get cycle_step_based variables:
    if len(keys_cyc): 
        for key in keys_cyc:
            if key in ["Discharge capacity [A.h]"]:
                step_no = step_CD;
                my_dict[key].append  (
                    Sol.cycles[cycle_no].steps[step_no][key].entries[-1]
                    - 
                    Sol.cycles[cycle_no].steps[step_no][key].entries[0])
            elif key[0:5] in ["CDend","CCend","CVend","REend",
                            "CDsta","CCsta","CVsta","REsta",]:
                step_no = eval("step_{}".format(key[0:2]))
                if key[2:5] == "sta":
                    my_dict[key].append  (
                        Sol.cycles[cycle_no].steps[step_no][key[6:]].entries[0])
                elif key[2:5] == "end":
                    my_dict[key].append  (
                        Sol.cycles[cycle_no].steps[step_no][key[6:]].entries[-1])
                
    # get location_based variables:
    if len(keys_loc): 
        for key in keys_loc:
            if key in ["x_n [m]","x [m]","x_s [m]","x_p [m]"]:
                #print("These variables only add once")
                if not len(my_dict[key]):   # special: add only once
                    my_dict[key] = Sol[key].entries[:,-1]
            elif key[0:5] in ["CDend","CCend","CVend","REend",
                            "CDsta","CCsta","CVsta","REsta",]:      
                #print("These variables add multiple times")
                step_no = eval("step_{}".format(key[0:2]))
                if key[2:5] == "sta":
                    my_dict[key].append  (
                        Sol.cycles[cycle_no].steps[step_no][key[6:]].entries[:,0])
                elif key[2:5] == "end":
                    my_dict[key].append  (
                        Sol.cycles[cycle_no].steps[step_no][key[6:]].entries[:,-1])
    return my_dict    
```

```python
# from Jianbo: 如何递归一个字典
#!/usr/bin/python3

'''
递归函数
- kvs 要扫描的字典
- key_list 递归函数里一般都有这么个参数，表示递归到哪里了，什么时候结束，
  有点像for循环里的i，用于推进和结束递归
- acc 递归函数里常见的参数，表示一路递归到当前位置的累计结果，
  在递归结束的时候，acc的值就是我们要求的结果之一
'''
def recursive_scan(kvs, key_list, acc):
    # 递归终止条件
    if len(key_list) == 0:
        print(acc)
        return

    # 继续递归
    k = key_list[0]
    for v in kvs[k]:
        acc[k] = v
        recursive_scan(kvs, key_list[1:], acc)


def scan(kvs):
    key_list = list(kvs.keys())
    recursive_scan(kvs, key_list, {})



if __name__ == '__main__':
    kvs = {
        "A": [11, 12, 13],
        "B": [21, 22],
        "C": [31, 32, 33, 34]
    }
    scan(kvs)

# 我的调用：
import copy
def recursive_scan(mylist,kvs, key_list, acc):
    # 递归终止条件
    if len(key_list) == 0:
        mylist.append(acc.copy())   
        # 注意这里不能直接写acc,否则只是引用了这个对象，并没有在内存上生成一个新东西，
        # 详细参考：https://www.runoob.com/w3cnote/python-understanding-dict-copy-shallow-or-deep.html copy.deepcopy(acc) 
        # 如果value是个list，就要deep copy了
        return mylist
    # 继续递归
    k = key_list[0]
    for v in kvs[k]:
        acc[k] = v
        # print(v)
        recursive_scan(mylist,kvs, key_list[1:], acc)
""" the following code is to test the function
kvs = {
    "A": [11, 12, 13],
    "B": [21, 22],
    "C": [31, 32, 33, 34]}

key_list = list(kvs.keys())
mylist = [];
recursive_scan(mylist,kvs, key_list, {}) """
```
